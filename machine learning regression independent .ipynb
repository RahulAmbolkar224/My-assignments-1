```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
sns.get_dataset_names()
```

```python
mpg = sns.load_dataset('mpg')
mpg.head()
```

```python
df= mpg.copy()
df.head()
```

```python
df.columns
```

```python
df.shape
```

```python
sns.regplot(x='horsepower',y='mpg',data=df)
plt.title("mpg vs horsepower trend")
```

```python
df['horsepower']= df['horsepower'].fillna(0)
```

```python
df[['horsepower','mpg']].corr()
```

```python
x = df['horsepower']
x = x.values.reshape(-1,1)
x
```

```python
y = df['mpg']
y = y.values.reshape(-1,1)
y
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model=LinearRegression()
model_train = model.fit(x,y)
print("The model Training is completed")
```

```python
model_train.coef_
```

```python
model_train.intercept_
```

```python
model_train.score(x,y)
```

```python
x1 = 250
```

```python
y =(model_train.coef_ * x1) + model_train.intercept_
y
```
